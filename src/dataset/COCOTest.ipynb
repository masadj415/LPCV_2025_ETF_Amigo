```python
from torchvision import transforms
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import os
import sys

# ROOT_DIR = os.path.abspath(os.path.join(os.getcwd(), ".."))
# sys.path.append(ROOT_DIR)

from DatasetReader import COCODataset
import utils as dsutils

all_classes = [s.lower() for s in dsutils.GLOBAL_CLASSES]

def get_coco_data_loader():
    transform = transforms.Compose([
        transforms.ToTensor(),  # Converts PIL.Image to a Tensor
        # transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    datasetCOCO = COCODataset(
        annotation_file=r"/home/centar15-desktop1/LPCV_2025_T1/datasets/coco/annotations/instances_train2017.json",
        image_dir=r'/home/centar15-desktop1/LPCV_2025_T1/datasets/coco/train2017',
        target_classes= ['spoon'],
        transform=transform,
        min_size = 70,
        max_expand=2
    )

    dataloader = DataLoader(datasetCOCO, batch_size=1, shuffle=True)
    return datasetCOCO, dataloader

dataset, dataloader = get_coco_data_loader()
def custom_collate_fn(batch):
    images, labels = zip(*batch)  # Unpack the batch into images and labels
    images = [transforms.ToTensor()(img) for img in images]  # Convert PIL images to tensors
    return images, labels


dataset, dataloader = get_coco_data_loader()




```
